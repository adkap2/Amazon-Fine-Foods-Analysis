```python
import pandas as pd
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
%config Completer.use_jedi = False
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
nltk.download('stopwords')
nltk.download('punkt')
import string
from ast import literal_eval
```

```python
data = pd.read_csv('preproc_Reviews.csv')
data['CleanText'] = data['CleanText'].apply(lambda x: literal_eval(x)) #need to parse lists back into lists
```

# Do preproc on summary

```python
data['CleanSummary'] = data['Summary']
```

```python
data['CleanSummary'] = data['CleanSummary'].apply(lambda x: str(x))
```

```python
data['CleanSummary'] = data['CleanSummary'].apply(lambda x: x.lower())
```

```python
data['CleanSummary'] = data['CleanSummary'].apply(lambda x: "".join([char for char in x if char not in string.punctuation]))
```

```python
data['CleanSummary'] = data['CleanSummary'].apply(lambda x: nltk.word_tokenize(x))
```

```python
stop_words = stopwords.words('english')
data['CleanSummary'] = data['CleanSummary'].apply(lambda x: [word for word in x if word not in stop_words])
```

```python
porter = PorterStemmer()
data['CleanSummary'] = data['CleanSummary'].apply(lambda x: [porter.stem(word) for word in x])
```

```python
data['CleanAll'] = data['CleanSummary'] + data['CleanText']
```

```python
data['CleanSet'] = data['CleanAll'].apply(lambda x: {word for word in x})
```

```python
data.to_csv('preproc_Reviews.csv')
```
